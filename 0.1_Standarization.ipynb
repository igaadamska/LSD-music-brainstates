# Standardization #
This notebook is used for standarization of the fMRI dataset (https://openneuro.org/datasets/ds003059/versions/1.0.0).

**Last update:** 17.03.2021

### Step 0: Loading libraries

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import sys
import os

from nilearn.image import clean_img #Main denoising function 
import nibabel as nb
```

### Step 2: Selecting data

```python
#Input path
data_path = '/home/iga/BrainStates/ds003059/data'
subjects = ['001','002','003','004','006','009','010','011','012','013','015','017','018','019','020']
sessions = ['PLCB', 'LSD']
task = 'rest'
runs = ['01','02','03']
suffix = 'nii.gz'

#Output path
output = '/home/iga/BrainStates/ds003059/data/standardized'
```

```python
#Iterationg over sessions
for ses in sessions:
    
    #Iterating over subjects
    for i, sub in enumerate(subjects):
        
        #Iterating over runs
        for k, run in enumerate(runs):
                
            print(f'Performig standarization for subject {sub} session {ses} run {run}')
            
            #Selecting denoised data
            subject_path = f'{data_path}/'
            file = f'sub-{sub}_ses-{ses}_task-{task}_run-{run}_bold.{suffix}'
            file_path = f'{subject_path}{file}'

            #Signal extraction
            fmri_standardized= clean_img(file_path, detrend=True, standardize=True)
            fmri_standardized_fname = f'standardized_{file}'
            nb.save(fmri_standardized, os.path.join(output, fmri_standardized_fname))
            
            print(f'Saving {fmri_standardized_fname}')
```
